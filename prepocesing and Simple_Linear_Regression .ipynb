# Simple LinearRegression 
  Formula: y=mx+c
  x is the the slope 
  y is the intercept or constant 
  Inorder to assign a constant to Simple LinearRegression we use the 
  ## Import statsmodel.api 
      1.Above command is used to create or assing the constant variable to linear equation 
      2.Used to execute the simple_linear_regression  equation
      
      

```python
import statsmodels.api as sn
```

```python
X=sn.add_constant(df['room_num']) #constant variable
```

```python
lm=sn.OLS(df["price"],X).fit()   #puttinf x and y variables in the Simple_linear_regression equation
```

```python
lm.summary()    #summary() means it gives the results
```

# Ridge and Lasso Regression 
Before doing ridge and lasso one should standalize the data
to standalize the data we need a package

```python
from sklearn import preprocessing
```

Scaler is used to store the infromation about how to store the columns

Standardizing the independent variables

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha =0.5)
lm_r.fit(X_train_s, y_train)
```

```python
r2_score(y_test,lm_r.predict(X_test_s))
```
